# Week 3 Web Scraper

This should be relatively straightforward to interpret. Each table is pretty-printed with describe() and I've called out each homework objective in a print statement.

I've opted to include all the requests in a single cell to make the code less verbose. Also note that the data on the table changes - sometimes `NewCases` is parseable as a good numeric value, and sometimes it is `NaN` - I think that pandas sometimes gets confused with `+/-` characters applied to the numeric string. 

```python
import pandas as pd
import requests

# I get a 400 if I try to run this directly
request_obj = requests.get("https://www.worldometers.info/coronavirus/country/us/#nav-yesterday")
covid_dataframe = pd.read_html(request_obj.text)[0]

# drop first and last column
covid_dataframe.drop(covid_dataframe.columns[[0, -1]], inplace=True, axis=1)

# drop 'totals' rows
covid_dataframe.drop(0, inplace=True)
covid_dataframe.drop(covid_dataframe.tail(1).index, inplace=True)

print()
print("Printing the first few rows of the entire frame")
print()
# pretty print the first few rows 
display(covid_dataframe.head())

print()
print("Summary statistics")
print()
# display summary statistics of the data
display(covid_dataframe.describe())

print()
print("Top 5 states by total deaths")
print()
sorted_by_total_deaths = covid_dataframe.sort_values("TotalDeaths", ascending=False)
display(sorted_by_total_deaths[["USAState", "TotalDeaths"]].head(n=5))

print()
print("Top 5 states by new cases")
print()
sorted_by_new_cases = covid_dataframe.sort_values("NewCases", ascending=False)
display(sorted_by_total_deaths[["USAState", "NewCases"]].head(n=5))

print()
print("Top 5 states by Tot Cases/1M pop")
print()
sorted_by_tot_cases_1m = covid_dataframe.sort_values("Tot Cases/1M pop", ascending=False)
display(sorted_by_tot_cases_1m[["USAState", "Tot Cases/1M pop"]].head())


print()
print("Top 5 states by Deaths/1M pop")
print()
sorted_by_total_deaths_1m = covid_dataframe.sort_values("Deaths/1M pop", ascending=False)
display(sorted_by_total_deaths_1m[["USAState", "Deaths/1M pop"]].head())
```
